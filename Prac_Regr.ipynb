```python
import scipy.stats as st
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
import pandas as pd

from sklearn.metrics import r2_score
import seaborn as sns
```

# Задача № 1

Сгенерировать Y = P(x) + $\varepsilon$

а) P(x) - многочлен 3 степени;

б) P(x) - многочлен 6 степени;

$x \sim U[0, 1], \ \varepsilon \sim N(0, \sigma^2)$

i) $\sigma = 0.2$; 
ii) $\sigma = 1$;

Построить полиномиальную модель 2, 3, 6 степени.

# a) Многочлен 3 степени

```python
x_size = 100

x = np.array(sorted(st.uniform.rvs(size=x_size)))

real_coeffs = [5, - 8.25, 3.825, -0.475] # 5 * (x - 0.2) * (x - 0.5) * (x - 0.95)

def P(x, coeffs):
    res = 0
    n = len(coeffs) - 1
    for i in range(len(coeffs)):
        res += x**(n - i) * coeffs[i]
    return res

p = np.vectorize(lambda x: P(x, real_coeffs))

y_clean = p(x)
```

```python
sigma = 0.05
eps = st.norm.rvs(scale=sigma, size=x_size)
y_dirty = y_clean + eps
```

```python
power_of_est_model = 2
X = np.stack([x**(power_of_est_model - i) for i in range(power_of_est_model)]).T

reg = LinearRegression().fit(X, y_dirty)

print(f'Estimated coeffs: {list(reg.coef_) + [reg.intercept_]}')
print(f'Real coeffs: {real_coeffs}')

fitted_value = reg.predict(X)

plt.plot(x, fitted_value)
plt.scatter(x, y_dirty, s=1)
plt.plot(x, y_clean)
plt.legend(['Fitted Graph', 'Sample', 'Original Graph'])
plt.title(f'$\sigma=${sigma},\n real model power = {len(real_coeffs) - 1}, \n estimation model power = {power_of_est_model}, \
           \n MSE = {np.linalg.norm(y_clean-fitted_value)**2 / len(y_clean)}')
plt.show()
```

# Задача № 2.

```python
def plot_reg_n_con(X, real_coeffs, eps):
    
    y_clean = list(map(lambda x: real_coeffs[0] * x[0] + real_coeffs[1], X))

    y_dirty = y_clean + eps

    reg = LinearRegression().fit(X, y_dirty)

    fitted_value = reg.predict(X)
    print(f'Estimated coeffs: {list(reg.coef_) + [reg.intercept_]}')
    print(f'Real coeffs: {real_coeffs}')

    RSS = np.sum((y_dirty - fitted_value)**2)
    n = len(X)
    k = 2

    q1 = st.t.ppf(df=n-k, q=0.05 / 2)
    q2 = st.t.isf(df=n-k, q=0.05 / 2)

    x = X.flatten()

    x_ = np.stack([x, np.ones(len(x))])
    x_matr = np.linalg.inv(x_.dot(x_.T))
    
    denom = np.sqrt(RSS * (1 + x_.T.dot(x_matr).dot(x_).diagonal()) / (n - k))
    
    fig, ax = plt.subplots()


    plt.fill_between(x, fitted_value - q2 * denom, fitted_value - q1 * denom, alpha=0.5)
    plt.plot(x, fitted_value)
    plt.plot(x, y_clean)
    plt.legend(['confidense interval', 'fitted value', 'real function'])
    plt.show()
    
```

```python
sample_size = 30
X = np.array(sorted(st.uniform.rvs(size=sample_size))).reshape(-1, 1)
real_coeffs = [7, 2]
sigma = 1
# NORMAL ERROR
eps = st.norm.rvs(size=sample_size, scale=sigma)
plot_reg_n_con(X, real_coeffs, eps)
```

```python
sample_size = 30
X = np.array(sorted(st.uniform.rvs(size=sample_size))).reshape(-1, 1)
real_coeffs = [7, 2]
sigma = 1
# Laplace ERROR
eps = st.laplace.rvs(size=sample_size, scale=1/sigma)
plot_reg_n_con(X, real_coeffs, eps)
```

```python
sample_size = 30
X = np.array(sorted(st.uniform.rvs(size=sample_size))).reshape(-1, 1)
real_coeffs = [7, 2]

sigma = 1
# ANOTHER ERROR
eps1 = st.uniform.rvs(size=sample_size)
eps2 = st.uniform.rvs(size=sample_size)

eps = 1 / (1 - eps1) * (1*(eps2 <= 0.5) - 1*(eps2 > 0.5))

plot_reg_n_con(X, real_coeffs, eps)
```

```python
X = np.stack([df.Year]).T

reg = LinearRegression().fit(X, np.log(df.Price))

x = np.array(sorted(df.Year.unique()))

fitted_value = reg.predict(x.reshape(-1, 1))
print(f'Estimated coeffs: {list(reg.coef_) + [reg.intercept_]}')

plt.scatter(df.Year, np.log(df.Price), s=0.5)
plt.plot(x, fitted_value)
```

```python


fitted_value = reg.predict(np.array(df.Year).reshape(-1, 1))

residuals = np.log(df.Price) - fitted_value

model_norm_residuals_abs_sqrt=np.sqrt(np.abs(residuals))

plt.figure(figsize=(7,7))
sns.regplot(x=fitted_value, y=model_norm_residuals_abs_sqrt,
              scatter=True,
              lowess=True,
              line_kws={'color': 'red', 'lw': 1, 'alpha': 0.8})
plt.ylabel("Standarized residuals")
plt.xlabel("Fitted value")
```
