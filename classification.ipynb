```python
n = 200

s1 = np.diag([1, 1])
m1 = [0, 0]
x1 = st.multivariate_normal.rvs(size=n, mean=m1, cov=s1)

s2 = np.diag([1, 1])
m2 = [1, 1]
x2 = st.multivariate_normal.rvs(size=n, mean=m2, cov=s2)


plt.scatter(*x1.T, marker="o")
plt.scatter(*x2.T, marker="<")

X = np.vstack([x1, x2])
y = np.hstack([np.zeros(len(x1)), np.ones(len(x2))])


X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.5, shuffle=True)
```

```python
from sklearn.preprocessing import label_binarize
from sklearn.model_selection import train_test_split
from sklearn.multiclass import OneVsRestClassifier
from sklearn.metrics import precision_recall_curve
from sklearn.metrics import average_precision_score
from sklearn.metrics import PrecisionRecallDisplay
from sklearn.metrics import RocCurveDisplay
```

```python
df = pd.read_csv('./datasets/iris.txt')
df.head()
```

```python
enum_species = []

match = {
    'setosa': 0,
    'virginica': 1,
    'versicolor': 2
}

reversed_match = dict(zip(match.values(), match.keys()))

for el in df.species.to_numpy():
    enum_species.append(match[el])

df['enum'] = enum_species
```

```python
X = df[['sepal_length', 'sepal_width', 'petal_length', 'petal_width']]
y = df.enum

lda = LDA(n_components=2)
X_r2 = lda.fit(X, y).transform(X)


colors = ["navy", "turquoise", "darkorange"]

plt.figure()
for color, i, target_name in zip(colors, match.values(), match.keys()):
    plt.scatter(
        X_r2[y == i, 0], X_r2[y == i, 1], alpha=0.8, color=color, label=target_name
    )
plt.legend(loc="best", shadow=False, scatterpoints=1)
plt.title("LDA of IRIS dataset")
```

# LDA

## One vs One

```python
labels = [0, 1]

X_train, X_test, Y_train, Y_test = train_test_split(
    df[df.enum.isin(labels)][['sepal_length', 'sepal_width', 'petal_length', 'petal_width']], 
    df[df.enum.isin(labels)].enum, 
    test_size=0.5,
    random_state=42
)

classifier = LDA()
classifier.fit(X_train, Y_train)

print(classification_report(Y_test, classifier.predict(X_test), labels=labels))

y_score = classifier.decision_function(X_test)
display = PrecisionRecallDisplay.from_predictions(1 - 1 * (Y_test == labels[0]), y_score, 
                                                  name=f'{reversed_match[labels[0]]} vs {reversed_match[labels[1]]}')


RocCurveDisplay.from_predictions(
    1 - 1 * (Y_test == labels[0]),
    y_score,
    color="darkorange",
    name=f'{reversed_match[labels[0]]} vs {reversed_match[labels[1]]}'
)

plt.axis("square")
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.legend()
plt.show()
```

## One vs Other

```python
X = df[['sepal_length', 'sepal_width', 'petal_length', 'petal_width']]
Y = label_binarize(df.enum, classes=[0, 1, 2])
n_classes = Y.shape[1]

X_train, X_test, Y_train, Y_test = train_test_split(
    X, Y, test_size=0.5, random_state=42
)

classifier = OneVsRestClassifier(
    LDA()
)
classifier.fit(X_train, Y_train)
y_score = classifier.decision_function(X_test)
```

```python
precision = dict()
recall = dict()

for i in range(n_classes):
    precision[i], recall[i], _ = precision_recall_curve(Y_test[:, i], y_score[:, i])

precision["micro"], recall["micro"], _ = precision_recall_curve(
    Y_test.ravel(), y_score.ravel()
)

colors = ["navy", "turquoise", "darkorange", "cornflowerblue", "teal"]
_, ax = plt.subplots(figsize=(7, 8))
for i, color in zip(range(n_classes), colors):
    display = PrecisionRecallDisplay(
        recall=recall[i],
        precision=precision[i]
    )
    display.plot(ax=ax, name=f"Precision-recall for class {reversed_match[i]}", color=color)

display = PrecisionRecallDisplay(
    recall=recall["micro"],
    precision=precision["micro"]
)
display.plot(ax=ax, name="Micro-average precision-recall", color="gold")
```

```python
colors = ["navy", "turquoise", "darkorange", "cornflowerblue", "teal"]

_, ax = plt.subplots(figsize=(7, 8))

for i, color in zip(range(n_classes), colors):
    display = RocCurveDisplay.from_predictions(
        Y_test[:, i],
        y_score[:, i],
        ax=ax, 
        name=f"ROC-AUC for class {reversed_match[i]}", 
        color=color
    )

RocCurveDisplay.from_predictions(
    Y_test.ravel(),
    y_score.ravel(),
    name=f"micro",
    color=color,
    ax=ax
)
```

```python
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.ensemble import RandomForestClassifier
```

```python
df = pd.read_csv('./datasets/spam.txt', sep=' ', header=None)
df.head()
```

```python
X = df.drop(columns=[57])
y = df[57]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
classifier = DecisionTreeClassifier(max_depth=4, min_samples_leaf=5).fit(X=X_train, y=y_train)

print(classification_report(y_test, classifier.predict(X_test), labels=[0, 1]))

display = PrecisionRecallDisplay.from_predictions(y_true=y_test, y_pred=classifier.predict(X_test))


RocCurveDisplay.from_predictions(
    y_true=y_test, 
    y_pred=classifier.predict(X_test),
    color="darkorange"
)

plt.axis("square")
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.legend()
plt.show()
```

```python
X.iloc[:, classifier.feature_importances_ == classifier.feature_importances_.max()].columns
```

```python
plt.figure(figsize=(20, 10))
plot_tree(classifier, fontsize=5)
plt.show()
```

```python
classifier = RandomForestClassifier(max_depth=5, n_estimators=200, min_samples_leaf=5).fit(X=X_train, y=y_train)

print(classification_report(y_test, classifier.predict(X_test), labels=[0, 1]))

display = PrecisionRecallDisplay.from_predictions(y_true=y_test, y_pred=classifier.predict(X_test))


RocCurveDisplay.from_predictions(
    y_true=y_test, 
    y_pred=classifier.predict(X_test),
    color="darkorange"
)

plt.axis("square")
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.legend()
plt.show()
```
