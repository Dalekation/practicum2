```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn import datasets
from sklearn.decomposition import PCA
```

```python
iris = datasets.load_iris()
y = iris.target

X_reduced = PCA(n_components=4).fit(iris.data)
```

```python
X_reduced.explained_variance_ratio_, X_reduced.explained_variance_ratio_.sum()
```

```python
X_reduced.components_
```

```python
variance = X_reduced.explained_variance_ratio_.copy()

for i in range(1, 4):
    variance[i] = variance[i] + variance[i-1]

plt.plot(range(1, 5), variance)
plt.title('Variance ratio growth')
```

```python
X_reduced.components_, X_reduced.singular_values_
```

```python
iris.feature_names
```

```python
values = X_reduced.singular_values_.copy() / X_reduced.singular_values_.sum()

for i in range(1, 4):
    values[i] = values[i] + values[i-1]

plt.plot(range(1, 5), values)
plt.title('Singular values growth')
```

```python

fig = plt.figure(1, figsize=(8, 6))
ax = fig.add_subplot(111, projection="3d", elev=-150, azim=110)
ax.scatter(
    iris.data[:, 0],
    iris.data[:, 1],
    iris.data[:, 2],
    c=y,
    cmap=plt.cm.Set1,
    edgecolor="k",
    s=40,
)
```

```python
iris.feature_names
```

```python
X_reduced = X_reduced.transform(iris.data)
fig = plt.figure(1, figsize=(8, 6))
ax = fig.add_subplot(111, projection="3d", elev=-150, azim=110)
ax.scatter(
    X_reduced[:, 0],
    X_reduced[:, 1],
    X_reduced[:, 2],
    c=y,
    cmap=plt.cm.Set1,
    edgecolor="k",
    s=40,
)
```
